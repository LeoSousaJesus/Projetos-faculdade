# Carregando um DF de dados existentes

Os DataFrames podem ser carregados a partir de uma diversidade de tipos de dados. Para o nosso exemplo, carregaremos dados a partir de um arquivo CSV.

Esse carregamento pode ser feito a partir de uma pasta local, utilizando como endereço de entrada a localização física do arquivo, ou a partir de uma url válida.

```python
url = 'https://raw.githubusercontent.com/betotorres/datasets/main/ds_preco_veiculos.csv'
preco_df = pd.read_csv(url, sep=',')
preco_df.head(2)
```

```python
preco_df.info()
```

# Filtrando registros no DF
O DF permite, além de uma variedade de outras formas, a filtragem de colunas usando as operações básicas de comparação, a exemplo de <, >, =, e as conjunções lógicas e (&) e ou (!).

```python
preco_df[preco_df['seguranca'] == 3].head(2)
```

```python
preco_df[((preco_df['seguranca'] >= 3) & (preco_df['tipo_combustivel'] == 'gas')) | (preco_df['rpm_max'] > 5000)].head(2)
```
