#Biblioteca Pandas

Pandas é uma biblioteca de manipulação e análise de dados de código aberto para Python. Desenvolvida por Wes McKinney, o Pandas fornece estruturas de dados de alto desempenho e fáceis de usar e ferramentas de análise de dados que são essenciais para trabalhar com dados estruturados. Dentre as funcionalidades chaves do Pandas, destacam-se:

*   **Entrada/Saída de Dados** - suporta leitura e gravação de dados em vários formatos, incluindo CSV, Excel, bancos de dados SQL, JSON e muito mais.
*   **Limpeza de Dados** - fornece funções para lidar com dados ausentes, valores duplicados e outliers, dentre outros.
*   **Indexação e Seleção** - oferece diversas opções de indexação, permitindo que os usuários selecionem e manipulem dados com base em rótulos, posições ou condições booleanas.
*  **Agrupamento e Agregação** - permite o agrupamento de dados com base em uma ou mais colunas e a execução de funções de agregação, a exemplo de soma, média e contagem.

Quando já instalada, a biblioteca pandas pode ser instanciada conforme apresentado no comando abaixo:


```python
#Comando de importação da biblioteca pandas
import pandas as pd
```

#Pandas DataFrames

No núcleo do Pandas estão as suas duas principais estruturas de dados, as Series e os DataFrames. As Series são uma  matriz rotulada unidimensional capaz de conter qualquer tipo de dados, como se fosse uma coluna em uma planilha, mas com flexibilidade e funcionalidade adicionais. A estrutura de dados mais icônica dos Pandas, entretanto, é o DataFrame.

O DataFrame é uma estrutura de dados tabular bidimensional com eixos rotulados  tendo formato similar ao de uma tabela convencional, vez que é composto por linhas e colunas. Os Dataframe oferecem ao desenvolvedor uma interface semelhante a uma planilha eletrônica, facilitando a manipulação e a análise de dados. Eles permitem a integração perfeita de vários tipos de dados e fontes, fornecendo um contêiner versátil para diversos conjuntos de dados.

São muitas as formas de se instanciar um Dataframe. Um dos métodos mais comuns é utilizar a estrutura de dicionários do python. Neste sentido, cada chave do dicionário irá representar uma coluna do Dataframe. Os valores associados a cada chave, por sua vez, podem ser declarados como listas e devem possuir a mesma quantidade de elementos, já que cada posição na lista representa uma linha do Dataframe.

```python
"""
Criação de um dicionário em python de nome "notas" e com três colunas - nome, idade e nota_matematica
- Valores de nome - João, José, Maria, Joana, Tiago, Júlia
- Valores de idade - 19, 22, 23, 21, 18, 19
- Valores de nota_matematica - 10, 9, 2, 9, 3, 8
"""
notas = {
    "nome":['Joao', 'José', 'Maria', 'Joana', 'Tiago', 'Julia'],
    "idade":[19,22, 23, 21, 18, 19],
    "nota_matematica":[10, 9, 2, 9, 3, 8]
}

#Instanciação do Dataframe utilizando o dicionário
notas_df = pd.DataFrame(notas)
```

## Explorando os Dados com head e tail

Ao trabalhar com grandes conjuntos de dados no Pandas, muitas vezes é impraticável examinar todo o conjunto de dados de uma só vez. É aqui que entram os métodos **head** e **tail**, que oferecm uma maneira rápida e concisa de observar os registros localizados no início e no fim do conjunto de dados. De forma prática, esses comandos se mostram muito úteis para:

* **Validação de Dados**: Antes de mergulhar profundamente na análise, use head e tail para validar se o seu DataFrame foi carregado corretamente e se os dados estão como esperado.

* **Resumos Rápidos**: Use head e tail em conjunto para obter um resumo conciso de ambas as extremidades do seu conjunto de dados, dando a você uma visão holística.


O método **head** permite que sejam visualizadas as primeiras linhas do DataFrame, fornecendo um instantâneo da estrutura e do conteúdo dos dados. Por padrão, ele exibe as primeiras cinco linhas, mas é possível especificar um número diferente de linhas a serem apresentadas.

```python
#Apresente as 5 primeiras linhas do DF
notas_df.head()
```

```python
#Apresente as 2 primeiras linhas do DF
notas_df.head(2)
```

Por outro lado, o método de **tail** fornece uma visão das últimas linhas do seu DataFrame. Semelhante ao head, o padrão mostra as últimas cinco linhas, mas pode ser personalizado para exibir um número diferente.

```python
#Apresente as 5 ultimas linhas do DF
notas_df.tail()
```

```python
#Apresente as 2 ultims linhas do DF
notas_df.tail(2)
```

#Manipulação de Colunas

No Pandas, as colunas são os blocos de construção do DataFrames. Desta forma, a manipulação eficaz de colunas é crucial para a análise de dados. São várias as técnicas para selecionar, indexar e manipular colunas,inicialmente, vamos analisar a seleção de colunas através dos seus rótulos.

No pandas, são duas as principais notações para acesso as colunas, uma utilizando o ponto, e uma segunda utilizando colchetes. Quando da utilização de colchetes e possível, ainda, selecionar múltiplas colunas de uma única vez.


```python
#Seleção de colunas utilizando ponto
notas_df.nome
```

```python
#Seleção de colunas utilizando colchetes
notas_df['nome']
```

```python
#Seleção de múltiplas colunas utilizando colchetes
notas_df[['nome', 'idade']]
```

Além da seleção, o pandas permite a execução das demais operações de manipulação de colunas, a exemplo da remoção, adição ou alteração do seu rótulo. A adição simples de uma nova coluna pode ser realizada com uma operação de atribuição relacionada a um rótulo ainda não utilizado no DataFrame.

```python
#Adição de uma nova coluna chamada ano_matricula com todos os valores definidos como 2024
notas_df['ano_matricula'] = 2024
notas_df
```

```python
"""
Adição de una nova coluna de nome cidade_nascimento a partir de um dicionário
com 4 linhas, de valores - Brasília, Salvador, São Paulo, Rio de Janeiro,
Brasília, Salvador
"""

notas_df['cidade_nascimento'] = ['Brasília', 'Salvador', 'São Paulo', 'Rio de Janeiro', 'Brasília', 'Salvador']
notas_df
```

A remoção de colunas, por sua vez, ocorre com a utilização do comando drop em conjunto com o nome das colunas a serem removidas.

```python
#Remoção da Coluna Ano Matricula
notas_df = notas_df.drop(columns=['ano_matricula'])
notas_df
```

Já para renomear uma ou múltiplas colunas, usa-se a sintaxe apresentada a seguir.

```python
notas_df= notas_df.rename(columns={'cidade_nascimento':'cidade'})
notas_df
```

# Operações Matemáticas sobre Dataframes
Quando colunas possuem valores numéricos, as operações matemáticas sobre elas funcionam em semelhança as Series. Abaixo, apresentamos a sintaxe para execução das operações matemáticas básicas sobre as colunas do DataFrame notas_df - soma, subtração, multiplicação e divisão.

```python
#Soma 3 ao valor da coluna nota_matematica
notas_df["nota_matematica"] = notas_df["nota_matematica"]+ 3
notas_df
```

```python
#Subtrai 3 ao valor da coluna nota_matematica
notas_df["nota_matematica"] = notas_df["nota_matematica"] - 3
notas_df
```

```python
#Multiplica por 2 o valor de nota_matematica
notas_df["nota_matematica"] = notas_df["nota_matematica"]*(2)
notas_df
```

```python
#Divide por 2 o valor de nota_matematica
notas_df["nota_matematica"] = notas_df["nota_matematica"]/(2)
notas_df
```

É possível também realizar operações matemáticas entre as colunas, sendo que a operação é realizada entre os valores de cada linha, conforme apresentado a seguir.

```python
"""
Criação de nova coluna de nome razao com o valor da nota_matematica
dividida pela idade
"""
notas_df["razao"] = notas_df["nota_matematica"]/notas_df["idade"]
notas_df
```

# Filtrando registros no DF

O DF permite, além de uma variedade de outras formas, a filtragem de colunas usando as operações básicas de comparação, a exemplo de <, >, =, e as conjunções lógicas e (&) e ou (!).

```python
#Filtragem do DF notas_df cujo nome é igual a Joao
notas_df[notas_df['nome'] == 'Joao']
```

```python
#Filtragem do DF notas_df cuja idade é maior do que 19 e nota_matematica maior do que 4
notas_df[(notas_df['idade'] >  19) & (notas_df['nota_matematica'] > 4)]
```
